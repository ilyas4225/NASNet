# NASnet Large with imagenet wieghts

```python
import tensorflow as tf
from tensorflow.keras.datasets import cifar100
from tensorflow.keras.applications.nasnet import NASNetLarge, preprocess_input
from tensorflow.keras.layers import Dense, Flatten
from tensorflow.keras.models import Model
from tensorflow.keras.optimizers import SGD
from tensorflow.keras.utils import to_categorical
```

```python

# Load the CIFAR-100 dataset
(train_images, train_labels), (test_images, test_labels) = cifar100.load_data()
```

```python

# Preprocess the input images
train_images = preprocess_input(train_images)
test_images = preprocess_input(test_images)
```

```python
# Convert the labels to one-hot encoding
train_labels = to_categorical(train_labels, num_classes=100)
test_labels = to_categorical(test_labels, num_classes=100)
```

```python
# Load the pre-trained NASNetLarge model
base_model = NASNetLarge(weights='imagenet', include_top=False, input_shape=(32, 32, 3))

# Add new classification layers on top of the pre-trained model
x = base_model.output
x = Flatten()(x)
x = Dense(512, activation='relu')(x)
predictions = Dense(100, activation='softmax')(x)

# Create the final model
model = Model(inputs=base_model.input, outputs=predictions)

# Freeze the pre-trained layers of the model
for layer in base_model.layers:
    layer.trainable = False

# Compile the model
opt = SGD(lr=0.001)
model.compile(optimizer=opt, loss='categorical_crossentropy', metrics=['accuracy'])
```

# NGA_net

```python
# Import required packages
import numpy as np
from keras.datasets import cifar10
#from keras.utils import to_categorical
from nsga_net.nsga import NSGA

# Load CIFAR-10 dataset
(x_train, y_train), (x_test, y_test) = cifar10.load_data()

# Normalize input images
x_train = x_train.astype('float32') / 255
x_test = x_test.astype('float32') / 255

# Convert labels to categorical one-hot encoding
num_classes = 10
y_train = to_categorical(y_train, num_classes)
y_test = to_categorical(y_test, num_classes)

# Define NSGA-Net parameters
input_shape = (32, 32, 3)
output_shape = num_classes
pop_size = 100
num_generations = 50
mutation_prob = 0.2

# Define NSGA-Net model architecture search space
model_space = [
    {'num_filters': [16, 32, 64, 128], 'kernel_size': [3, 5, 7], 'pool_size': [2, 3]},
    {'num_filters': [16, 32, 64, 128], 'kernel_size': [3, 5, 7]},
    {'num_filters': [32, 64, 128], 'kernel_size': [3, 5, 7]},
    {'num_filters': [32, 64, 128], 'kernel_size': [3, 5], 'pool_size': [2, 3]}
]

# Define NSGA-Net fitness function
def fitness_func(individual):
    model, num_layers = individual
    model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
    history = model.fit(x_train, y_train, batch_size=64, epochs=10, validation_split=0.1)
    acc = history.history['val_accuracy'][-1]
    params = model.count_params()
    return [1-acc, params/1e6]

# Initialize NSGA-Net optimizer
nsga = NSGA(model_space, input_shape, output_shape, fitness_func, pop_size=pop_size, mutation_prob=mutation_prob)

# Perform NSGA-Net search
best_model, best_fitness = nsga.evolve(num_generations)

# Evaluate best model on test set
best_model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
test_loss, test_acc = best_model.evaluate(x_test, y_test, verbose=0)
print('Test accuracy:', test_acc)
```
